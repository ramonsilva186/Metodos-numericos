```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
def arco_T(x,n): 
    return ((-1)**n * ( (x**(2*n+1)) / (2*n+1)))
```

## Formula de Machin

```python
m = 4 * np.arctan(1/5) - np.arctan(1/239)
m
```

```python
machin = 0
lista_machin = []

for i in range(21):
    machin = 4 *(4 * arco_T((1/5),i) - arco_T((1/239),i))
    lista_machin.append(machin)
```

```python
for i in range(len(lista_machin)):
    print(f'{i} \t {lista_machin[i]}')
```

```python
x = np.arange(0,20)
```

```python
lista_machin = np.array(lista_machin)
erro = np.abs(np.diff(lista_machin)/lista_machin[0:20])
```

```python
plt.figure(figsize = (12, 8))
plt.plot(x, erro, 'black', label = "Machin")
plt.title("Formula de Machin")
plt.xticks(x, x)
plt.grid(ls=':')
plt.xlabel('$Termos$')
plt.ylabel('$f(x)$')
plt.legend()
```

## Formula de Hutton

```python
h = np.arctan(1/2) + np.arctan(1/3)
h
```

```python
hutton = 0
lista_hutton = []

for i in range(21):
    hutton = 4 * (arco_T(1/2,i) + arco_T(1/3,i) )
    lista_hutton.append(hutton)
```

```python
for i in range(len(lista_hutton)):
    print(f'{i} \t {lista_hutton[i]}')
```

```python
lista_hutton = np.array(lista_hutton)
erro_h = np.abs(np.diff(lista_hutton)/lista_hutton[0:20])
```

```python
plt.figure(figsize = (12, 8))
plt.plot(x, erro_h, 'green', label = "Hutton")
plt.title("Formula de Hutton")
plt.xticks(x, x)
plt.grid(ls=':')
plt.xlabel('$Termos$')
plt.ylabel('$f(x)$')
plt.legend()
```
